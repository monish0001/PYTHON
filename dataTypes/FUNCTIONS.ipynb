```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Sine Wave
ax1.plot(x, y1, color='blue', label='sin(x)')
ax1.set_title("Sine Wave")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
ax1.grid(True)
ax1.legend()

# Plot 2: Cosine Wave
ax2.plot(x, y2, color='red', linestyle='--', label='cos(x)')
ax2.set_title("Cosine Wave")
ax2.set_xlabel("X-axis")
ax2.set_ylabel("Y-axis")
ax2.grid(True)
ax2.legend()

# Save the figure
plt.savefig("sine_cosine.png", dpi=300)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set theme
sns.set_theme(style="whitegrid")

# Create a DataFrame
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': np.random.randint(10, 100, size=5)
})

# Create a bar plot
sns.barplot(data=df, x='Category', y='Values', palette="viridis")
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()

# Create a heatmap
data = np.random.rand(5, 5)
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.title("Heatmap Example")
plt.show()

# Create a pair plot
df = pd.DataFrame({
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10)
})
sns.pairplot(df)
plt.suptitle("Pair Plot Example", y=1.02)
plt.show()
```
